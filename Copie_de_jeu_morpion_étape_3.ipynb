```python

import ipywidgets as widgets
from IPython.display import display


# VARIABLES

chiffres = [1,2,3,4,5,6,7,8,9]
mark = ''
compteur = 0

cases = ["panneau"," "," "," "," "," "," "," "," "," "]


# ETIQUETTES DES JOUEURS

joueurX = widgets.HTML("<h3>Joueur 1 : X</h3>")
joueurO = widgets.HTML("<h3>Joueur 2 : O</h3>")

# RESULTAT

resultat = widgets.HTML("<h3>Tour du joueur X</h3>")


# FONCTION VICTOIRE


def victoire():

    combinaisons = [
        (1,2,3),
        (4,5,6),
        (7,8,9),
        (1,4,7),
        (2,5,8),
        (3,6,9),
        (1,5,9),
        (3,5,7)
    ]

    for a,b,c in combinaisons:

        if cases[a] == cases[b] == cases[c] and cases[a] != " ":
            return cases[a]

    return None


# FONCTION DE CLIC

def verifier(chiffre):

    global compteur
    global mark

    if chiffre in chiffres:

        chiffres.remove(chiffre)

        # Tour du joueur
        if compteur % 2 == 0:
            mark = "X"
        else:
            mark = "O"

        # Mise à jour du plateau
        cases[chiffre] = mark

        # Mise à jour du bouton
        boutons[chiffre-1].description = mark

        compteur += 1

        # Vérification victoire
        gagnant = victoire()

        if gagnant == "X":

            resultat.value = "<h2>Joueur 1 gagne</h2>"

        elif gagnant == "O":

            resultat.value = "<h2>Joueur 2 gagne</h2>"

        elif compteur > 8:

            resultat.value = "<h2>Match Égal</h2>"

        else:

            if compteur % 2 == 0:
                resultat.value = "<h3>Tour du joueur X</h3>"
            else:
                resultat.value = "<h3>Tour du joueur O</h3>"


# CREATION DES BOUTONS


boutons = []

for i in range(1,10):

    bouton = widgets.Button(
        description=" ",
        layout=widgets.Layout(width='70px', height='70px'))

    bouton.on_click(lambda b, x=i: verifier(x))

    boutons.append(bouton)


# ORGANISATION DU PLATEAU


grille = widgets.GridBox(
    boutons,
    layout=widgets.Layout(
        grid_template_columns="70px 70px 70px",
        grid_gap="5px" ))


# AFFICHAGE


display(widgets.HTML("<h1>MORPION</h1>"))

display(joueurX)
display(joueurO)

display(grille)

display(resultat)
```
