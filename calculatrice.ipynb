```python
# Définition de la fonction calculatrice
def calculatrice(num1, num2):
    # Demander l'opérateur à l'utilisateur
    operateur = input("Entrez un opérateur (+, -, *, /) : ")

    # Conditions pour choisir l'opération
    if operateur == "+":
        return num1 + num2

    elif operateur == "-":
        return num1 - num2

    elif operateur == "*":
        return num1 * num2

    elif operateur == "/":
        # Gestion de la division par zéro
        if num2 == 0:
            return "Erreur : division par zéro impossible !"
        return num1 / num2

    else:
        # Opérateur invalide
        return "Opérateur invalide ! Veuillez entrer +, -, * ou /"


# Programme principal avec boucle
while True :
    # Demander les nombres à l'utilisateur
    num1 = float(input("Entrez le premier nombre : "))
    num2 = float(input("Entrez le deuxième nombre : "))

    # Appeler la fonction calculatrice
    resultat = calculatrice(num1, num2)

    # Afficher le résultat
    print("Résultat :", resultat)

    # Demander si l'utilisateur veut continuer
    continuer = input("Voulez-vous continuer ? (oui/non) : ")
    if continuer.lower() != "oui":
        print("Merci d'avoir utilisé la calculatrice ")
        break
```
